```python
# ttps://github.com/dhrim/deep_learning_data 프로젝트의 eeg.csv를 다운로드

# - channel 컬럼의 종류가 몇 개인지 출력
# - 각 channel 별로 데이터가 몇 개 인지 출력
# - channel 0의 sensor value 컬럼 만 추려서 plt.plot()으로 그리기.
# - 하나의 그림에 각 channel별 sensor value 컬럼을 plt.plot()을 겹쳐서 그리기.
# - 각 채널 별로 plt.plot()로 한 장씩 그리기
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('eeg.csv')
```

```python
df
```

```python
# - channel 컬럼의 종류가 몇 개인지 출력
print(len(df.channel.unique()))
```

```python
# - 각 channel 별로 데이터가 몇 개 인지 출력
print(df.channel.value_counts())
```

```python
# - channel 0의 sensor value 컬럼 만 추려서 plt.plot()으로 그리기.
plt.plot(df['sensor value'][df.channel==0],color = 'c')
```

```python
# - 하나의 그림에 각 channel별 sensor value 컬럼을 plt.plot()을 겹쳐서 그리기.
for i in df.channel:
  plt.plot(df['sensor value'][df.channel==i])
plt.show()
```

### scatter() - 산점도 그래프 그리기 

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.scatter(x, y)
plt.show()
plt.plot(x,y)
plt.show()
```

```python
x = np.random.randn(30)
y = np.random.randn(30)
size = np.random.rand(30)*1000
color = np.random.rand(30)

# s : 점 크기 
# alpha : 투명도
# cmap : 컬러맵 
# c : 컬러맵 중 컬러 선택 

plt.scatter(x,
            y, 
            s = size,  
            alpha = 0.4, 
            cmap = 'viridis',
            c= color
            )
plt.colorbar() # 해당 plot에서 나타나는 컬러들의 바를 보여 줌 
```

```python
plt.scatter(x,
            y, 
            s = size,  
            alpha = 0.7, 
            cmap = 'inferno',
            c= y
            )
plt.colorbar() 
```

#### 그래프에 요소 추가하기 - 제목, 축 이름, 범례 

```python
fig = plt.figure() # 꾸미기 시작하는 선언문? 비스무리 
fig.suptitle('Sine and Cosine') #제목 
plt.plot(x,np.sin(x), label = 'sin')
plt.plot(x,np.cos(x), label = 'cos')

plt.xlabel('x label') #x축 이름 추가 
plt.ylabel('y label') #y축 이름 추가 
plt.legend() #범례 

plt.show()
```

```python
#그래프 요소 꾸미기 
fig = plt.figure() # 꾸미기 시작하는 선언문? 비스무리 
fig.suptitle('Sine and Cosine') #제목 
plt.plot(x,np.sin(x), label = 'sin')
plt.plot(x,np.cos(x), label = 'cos')

plt.xlabel('x label') #x축 이름 추가 
plt.ylabel('y label') #y축 이름 추가 
plt.legend(frameon = True, loc = 'upper left') #범례, frameon = True : 범례 표시에 사각형을 만들어 줌, loc : 범례의 위치 
#loc = upper left, upper right, lower left, lower right, center, right, best, center left, center right 
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
x = np.linspace(0,10,100)
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.plot(np.sin(x))

ax2 = fig.add_subplot(1,2,2)
ax2.plot(np.cos(x))

plt.show()
```

```python
x = np.linspace(0,10,100)

fig = plt.figure()
plt.suptitle('sin or cos')

ax1 = fig.add_subplot(2,2,1) #subplot() 이 괄호 안에 있는게 위치 지정 
ax1.plot(np.sin(x))


ax2 = fig.add_subplot(2,2,2)
ax2.plot(np.cos(x))
ax2.grid(False) # 격자 없앨 때 plt.grid와 마찬가지로 ax2.grid 로 사용 

plt.show()

x = np.linspace(0,10,100)
fig = plt.figure()

ax1 = fig.add_subplot(224) #이런 식으로 쉼표 없이 쓸 수도 있음 
ax1.plot(np.sin(x))

ax2 = fig.add_subplot(223)
ax2.plot(np.cos(x))

plt.show()
```

#### subplot에 제목 붙이기 

```python
x = np.linspace(0,10,100)

fig = plt.figure()
plt.suptitle('sin or cos')

ax1 = fig.add_subplot(2,2,1) #subplot() 이 괄호 안에 있는게 위치 지정 
ax1.plot(np.sin(x))
ax1.set_title('sin') #서브플랏 제목 붙이기 

ax2 = fig.add_subplot(2,2,2)
ax2.plot(np.cos(x))
ax2.grid(False) # 격자 없앨 때 plt.grid와 마찬가지로 ax2.grid 로 사용 
ax2.set_title('cos')

plt.show()

x = np.linspace(0,10,100)
fig = plt.figure()

ax1 = fig.add_subplot(224) #이런 식으로 쉼표 없이 쓸 수도 있음 
ax1.plot(np.sin(x))
ax1.set_title('sin')

ax2 = fig.add_subplot(223)
ax2.plot(np.cos(x))
ax2.set_title('cos')

plt.tight_layout() #그래프를 딱 맞게? 안겹치게? 그려줌 
plt.show()
```
