```python
import seaborn as sns 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#age 에 대한 histogram 그리기 
sns.histplot(data=titanic, x='age',bins=100)
plt.savefig('plot1.jpg')
plt.show()
```

```python
#- age 컬럼에 대한 survivied를 histogram으로 그리기. 하나의 나이에 대해 2개의 bar가 있도록
sns.histplot(data=titanic, x='age',multiple='dodge',hue = 'survived', shrink = 0.7)
plt.savefig('plot2.jpg')
plt.show()
```

```python
#age 컬럼에 대한 생존율을 비교하도록 multiple 옵션을 fill로 해서 그리기
sns.histplot(data=titanic, x='age',multiple='fill',hue = 'survived', shrink = 0.5)
plt.savefig('plot3.jpg')
```

```python
#- age 컬럼을 x로 fare 컬럼을 y로 해서 scatter 그림 그리기
sns.scatterplot (data=titanic,x='age',y='fare')
plt.savefig('plot4.jpeg') #이미지 저장 
```

### violinplot()

```python
sns.set_theme(style = 'whitegrid')
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(y=tips.tip)
plt.show() 
#바이올린 플랏과 박스플롯 쓰는 방식 거의 유사 
```

```python
sns.violinplot(y=tips.tip)
```

```python
plt.subplot(121)
sns.boxplot(y=tips.tip)

plt.subplot(122)
sns.violinplot(y=tips.tip)

plt.show()
```

```python
#x축 별로 여러개 plot 그리기 
plt.subplot(121)
sns.boxplot(data=tips, x='day',y='total_bill')


plt.subplot(122)
sns.violinplot(data=tips, x='day',y='total_bill')

plt.tight_layout() #두개 그래프 자리 띄우는 거 
plt.show()
```

#### 정렬 

```python
#x축 별로 여러개 plot 그리기 
plt.subplot(121)
sns.boxplot(data=tips, x='time',y='total_bill',order = ['Dinner','Launch'])


plt.subplot(122)
sns.violinplot(data=tips, x='time',y='total_bill',order = ['Dinner','Launch'])

plt.tight_layout() #두개 그래프 자리 띄우는 거 
plt.show()
```

```python
#가로 세로 바꾸기 
plt.subplot(121)
sns.violinplot(data=tips, y="time", x="total_bill", order=["Dinner", "Lunch"])


plt.subplot(122)
sns.violinplot(data=tips, x="time", y="total_bill", order=['Dinner','Lunch'])


plt.tight_layout() #두개 그래프 자리 띄우는 거 
plt.show()
```

```python
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker", split=True)

plt.tight_layout()
plt.show()
```

#### boxplot 스타일 바꾸기 
- default : 박스 
- quartile 


```python
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker", inner = "quartile")

plt.tight_layout()
plt.show()
```

### stripplot() 

```python
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill")

plt.subplot(122)
sns.stripplot(data=tips, x="day", y="total_bill", jitter=0.5) #점 다닥다닥 붙어잇는걸 보여주는게 jitter plot 

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill")
sns.stripplot(data=tips, x="day", y="total_bill", color = '0.4') #바이올린 플랏과 스트립 플랏 겹치기 

plt.subplot(122)
sns.stripplot(data=tips, x="day", y="total_bill", jitter=0.5) #

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.boxplot(data=tips, x="day", y="total_bill")
sns.stripplot(data=tips, x="day", y="total_bill", color = '0.4') #바이올린 플랏과 스트립 플랏 겹치기 

plt.subplot(122)
sns.stripplot(data=tips, x="day", y="total_bill")

plt.tight_layout()
plt.show()
```

### heatmap()

```python
import numpy as np 

a = np.random.rand(10,8)

sns.heatmap(a)
plt.show()
sns.histplot(a.flatten())
plt.show()
```

#### faltten() 

```python
#2차원 어레이를 1차원 어레이로 바꾼 뒤 히트맵 
t = np.random.rand(3,60)
# sns.histplot(t.reshape(180,),bins=30)
sns.histplot(t.flatten(),bins=30)#flatten() : 다차원 어레이를 1차원 어레이로 바꿔 줌 
plt.show()
```

#### 범위 설정 : vmin, vmax 

```python
import numpy as np 

plt.figure(figsize=(10,4))
a = np.random.rand(10,8)

plt.subplot(121)
sns.heatmap(a)

plt.subplot(122)
sns.heatmap(a, vmin=0.2,vmax=0.8)
```

#### linewidth , cmap, cbar 

```python
#linewidth
#cmap
#cbar

import numpy as np 

plt.figure(figsize=(10,4))
a = np.random.rand(10,8)

plt.subplot(121)
sns.heatmap(a,linewidth = 0.5) #격자에 라인  추가 

plt.subplot(122)
sns.heatmap(a, linewidth=0.5, cmap='Blues',cbar = False) 

plt.show()
```
