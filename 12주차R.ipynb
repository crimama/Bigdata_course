## ggplot2 패키지로 그래프 그리기 

```r
library('ggplot2')
library('ggthemes')

str(diamonds) #ggplot 안에 들어있는 데이터 프레임 
```

```r
#aes = 축값 설정 
#ggplot은 일종의 캔버스로 밑그림만 그려 줌 
# 실질적인 데이터 그림은 geom_point로 그릴 수 있음
# geom_point는 산점도 그리는 함수 
#마지막으로 theme_wsj로 스타일을 입혀줌 
a <- ggplot(data=diamonds, aes(x=carat, y=price, colour = clarity))+
    geom_point() + theme_wsj()
a #ggplot 은 객체에 담을 수 있음 
```

### ggplot 객체 합치기 

```r
g1 <- ggplot(data = df, aes(x=height, y=weight, colour =bloodtype))
g2 <- geom_point()

#g1 + g2 <- 이것도 가능 
# g1 + geom_line() #<- 이것도 가능 
# g1 + geom_line() + geom_point()
g1 + geom_line(size = 1) + geom_point(size = 5) #사이즈 넣기 
```

### colour =  추가 

```r
g1 + geom_line(aes(colour = sex)) + geom_point(size = 10) 
#geom_ 에 colour = 를 추가로 입력할 수 있는데, 이건 전에 것을 덮어 씌움 
#colour는 일종의 hue, 분류를 만들어 줌
#ggplot만이 아니라 geom에도 축을 만들어야 하지만 만들지 않으면 ggplot꺼를 이어서 가져 옴 
#그런데 geom에 축을 새로 만들면 덮어 씌움 
# 이 경우 geom_line에는 colour = sex로 새로운 분류를 만들어 주어서 전체 선은 성 별로 그려짐 
#그런데 geom_point의 경우 새로 colour을 해주지 않아서 geom_point는 ggplot의 혈액형 분류를 그대로 가져 옴 
```

### face_grid(~@)

```r
#facet_grid(~@)를 넣으면 @를 기준으로 그래프를 나눠 줌, @가 독립변수가 됨 
#여기서 경우 ~sex를 했으므로 남성과 여성을 나눠서 각각 보여 줌 
#반대로 sex~. 방식으로 하면 sex가 종속변수가 됨
#sex ~ 하고 변수 하나를 작성 해주어야 함, 이 경우 '.'을 찍어줌 
g1 + geom_line(size=1) + geom_point(size=10) + facet_grid(sex~.)
```

```r
#facet_grid 에 cales = "free"를 넣으면 y축이 해당 그래프들의 값에 맞게 바뀜 
#그래서 이 scale="free"는 해당 그래프 안에서 비교할 때 주로 사용하고 
#남자 - 여자간 처럼 다른 그래프간 비교할 떄는 y축이 동일해야 하므로 빼는 것이 낫다 
g1 + geom_line(size=1) + geom_point(size=10) + facet_grid(sex~.,scales='free')
```

### facet_wrap

```r
g1 + geom_line(size=1) + geom_point(size=10) + facet_wrap(~sex,scales='free')
#facet_wrap 은 양쪽 그래프를 완전히 각각 보고자 할 때 주로 사용 
#x축과 y축이 모두 재정의 됨 
```

### 차량 종류 별 실린더에 따른 배기량과 연비

```r
#한개의 축을 바꾸고 싶을 때 scale 사용 
g + geom_point() + facet_grid(class~cyl, scale = 'free')
```

#### facet_wrap 사용 

```r
g + geom_point() + facet_wrap(class~cyl,nrow=4)
#이런 식으로 컬럼과 로우의 숫자 조절 가능 
# g + geom_point() + facet_wrap(class~cyl,ncol=4)
#facet_wrap을 쓰면 그래프가 완전히 따로 따로 됨 
```

#### 투명도 

```r
#산점도가 겹쳐있으면 몇개인지 알기 어렵다, 그래서 alhap로 투명도를 주는 것이 좋다 
#그냥 보는 것은 facet_wrap, 비교는 facet_grid 
g + geom_point(alpha = 0.5) + facet_wrap(class~cyl)
```

```r
#고수준 함수 
#a형 몇명 b형 몇명 c형 몇명 d형 몇명 
#혈액형을 기준으로 키의 평균을 그래프로 그림 
#각 혈액형별 키의 평균 막대그래프 그리기
b_mean <- tapply(df$height,df$bloodtype,mean) #tapply(데이터,기준,함수)
b_mean
```

```r
# ggplot 그리기 위해선 데이터프레임 형태로 들어와야 함
library('reshape2') #이걸로 데이터프레임 형태로 변환 
df_mean <- melt(b_mean)
df_mean
```

#### 막대 그래프에서 y축 값 바꾸기 

```r
ggplot(data=df_mean, aes(x=Var1,y=value,fill=Var1)) + geom_bar(stat='identity') 
#막대그래프에서 y축을 다른 값으로 바꾸고 싶을 때 stat='identity' 사용
```
