#### 히스토그램 - kde 라인 

```python
#추세선 추가하기 

plt.figure(figsize=(10,3))
plt.subplot(121)
sns.histplot(tips, x="total_bill", kde=True) #kde = True하면 라인이 추가 됨 

plt.subplot(122)
sns.histplot(tips, x="tip", kde=True)

plt.show()
```

#### 히스토그램 - 조건별 plot 후 겹치기 

```python
#성별 겹치기 

plt.figure(figsize=(10,3))
plt.subplot(121)
sns.histplot(tips, x="total_bill", hue="sex") #hue = 'sex' 하면 남자 여자 별로 그려지고 겹쳐짐 

plt.subplot(122)
sns.histplot(tips, x="tip", hue="sex")

plt.show()
```

#### 히스토그램-모양 바꾸기 

```python
#성별 겹치기 

plt.figure(figsize=(10,3))
plt.subplot(121)
sns.histplot(tips, x="total_bill", hue="sex", element = 'step') # 그래프 모양 -> step 형식으로 바꿈 

plt.subplot(122)
sns.histplot(tips, x="tip", hue="sex", element = 'poly') #그래프 모양을 바꿈 - 다각형 

plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.subplot(121)
sns.histplot(tips, x="total_bill", hue="sex", shrink=0.6) #막대의 크기를 줄임 

plt.subplot(122)
sns.histplot(tips, x="tip", hue="sex")

plt.show()
```

#### 히스토그램 - y축 값 변경 

```python
plt.figure(figsize=(10,3))
plt.subplot(121)
sns.histplot(tips, x="total_bill", hue="sex", stat='probability') #stat를 하면 y축 값이 갯수가 아닌 비율로 바뀜 

plt.subplot(122)
sns.histplot(tips, x="tip", hue="sex",stat='count')

plt.show()
```

```python
# 성별이라는 조건 별로 그래프를 보여주는데 multiple을 사용하면 겹치지 않고 동시에 보여 줌 
# stack으로 하면 낮은게 아래, 높은게 위로 해서 쌓아서 보여 줌 
# fill로 하면 y축이 0~1을 기준으로 조건별 비율을 토대로 보여 줌 
plt.figure(figsize=(10,3))
plt.subplot(121)
sns.histplot(tips, x="total_bill", hue="sex",multiple='fill')

plt.subplot(122)
sns.histplot(tips, x="tip", hue="sex",multiple='stack')

plt.show()
```

### 시각화 실습 

```python
titanic = sns.load_dataset('titanic')
titanic
```

1. 성별에 따른 생존자 수 
 - 남 녀 막대 그래프 
 - 남  녀 비율 그래프 
2. 나이별 생존자 수 
3. pclass별 생존자 수

#### 성별 

```python
#남녀수 
sns.histplot(data=titanic,x='sex',shrink=0.5)
plt.show()
```

```python
#남녀 생존자, 사망자 수 
sns.countplot(data=titanic,x='sex',hue='survived')
```

```python
sns.histplot(data=titanic,x='sex',multiple='dodge',hue='survived')
```

```python
#남 녀 생존자 비율 
sns.histplot(data=titanic,x='sex',hue='survived',multiple='fill',shrink=0.5)
```

#### 객실등급 

```python
#객실 등급 별 생존자 수 
sns.histplot(data=titanic, x="pclass", hue="survived", multiple="dodge", shrink=0.5)
plt.show()
```

```python
#객실 등급 별 생존자 비율 
sns.histplot(data=titanic, x="pclass", hue="survived", multiple="fill", shrink=0.5)
plt.show()
```

#### 요금 별 생존자 비율
